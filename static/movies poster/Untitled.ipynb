```python
import pandas as pd
movies = pd.read_csv('movies - Copy.csv') # Your list of movies here
print(movies)
```

```python
from bs4 import BeautifulSoup
import requests
import os

for movie in movies:
    # Search for the movie poster on Google Images
    query = f'{movie} movie poster'
    url = f'https://www.google.com/search?q={query}&tbm=isch'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the first image result
    img = soup.find('img')
    img_url = img['src']
    
    # Download the image
    img_data = requests.get(img_url).content
    filename = f'{movie}.jpeg'
    with open(filename, 'wb') as f:
        f.write(img_data)
```
